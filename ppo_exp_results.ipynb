```python
import json
import matplotlib.pyplot as plt
```

```python
# Load results from JSON files
experiments = {
    "Standard PPO": "exp1_results.json",
    "PPO with Bayesian Optimization": "exp2_results.json",
    "PPO with Random Search": "exp3_results.json",
    "PPO with Grid Search": "exp4_results.json"
}

results = {}
for exp_name, file_name in experiments.items():
    with open(file_name, "r") as f:
        results[exp_name] = json.load(f)
```

```python
# Plot Total Rewards
plt.figure(figsize=(12, 8))
for exp_name, exp_results in results.items():
    plt.plot(exp_results["rewards"], label=f"{exp_name} Rewards")
plt.title("Total Reward per Episode")
plt.xlabel("Episodes")
plt.ylabel("Total Reward")
plt.legend()
plt.grid(True)
plt.show()

# Plot Losses
plt.figure(figsize=(12, 8))
for exp_name, exp_results in results.items():
    plt.plot(exp_results["losses"], label=f"{exp_name} Losses")
plt.title("Loss per Episode")
plt.xlabel("Episodes")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Plot Entropy Evolution
plt.figure(figsize=(12, 8))
for exp_name, exp_results in results.items():
    plt.plot(exp_results["entropy"], label=f"{exp_name} Entropy")
plt.title("Entropy Evolution")
plt.xlabel("Episodes")
plt.ylabel("Entropy")
plt.legend()
plt.grid(True)
plt.show()
```
